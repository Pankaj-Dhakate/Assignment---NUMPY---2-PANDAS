#### Write a function to find moving average in an array over a window:
Test it over [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150] and window of 3.

```python
def mov_avg(m,k):
    for i in range(len(m)-k+1):
        print(round(((m[i]+m[i+1]+m[i+2])/k),3))
        
l=[3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150]
n=len(l)
k=3
print("Moving average for the list=[3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150] for window of 3 elements is as below")
print("Number of moving averages for the window of 3 elements => ",n-k+1)
mov_avg(l,k)
```

```python
import pandas as pd

df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})
a = pd.Series([7, 2, 0, 3, 4, 2, 5, 0, 3, 4])
df = df.assign(Y=pd.DataFrame({'Y':(a.groupby(a.eq(0).cumsum().mask(a.eq(0))).cumcount() + 1).mask(a.eq(0), 0).tolist()}))
print("Step difference from to the previous zero (or the start of the Series) as below => in Column Y ") 
df
```

#### 3. Find the sum of the values in s for every Wednesday

```python
import pandas as pd
import numpy as np
yr=pd.date_range(start = '2015 Jan 1 12:00', end = '2015 Dec 31 23:00', freq = 'D', tz='Asia/Kolkata').to_series()
d=yr.dt.dayofweek
r=np.random.rand(len(yr))
df = pd.DataFrame({'Daynumber':d,'RandomNumber':r})
s=df.loc[(df['Daynumber']==2),'RandomNumber'].sum()    # Daynumber==2 is 'wednesday' 
print("The sum of the values in s for every Wednesday in year => ",s)
s                                                      # Sum of all random numbers 
```

#### 4. Average For each calendar month

```python
import pandas as pd
import numpy as np
yr=pd.date_range(start = '2015 Jan 1 12:00', end = '2015 Dec 31 23:00', freq = 'D', tz='Asia/Kolkata').to_series()
m=yr.dt.month
r=np.random.rand(len(yr))
df1 = pd.DataFrame({'Monthnumber':m,'RandomNumber':r})
#avg=df1.loc[(df1['Monthnumber']),'RandomNumber'].mean()
grouped = df1.groupby(['Monthnumber']).mean()
print("Monthwise Average of RandomNumbers as below")
grouped
```

#### 5.    For each group of four consecutive calendar months in s, find the date on which the highest value occurred.
